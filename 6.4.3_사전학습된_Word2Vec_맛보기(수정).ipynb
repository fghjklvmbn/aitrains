# 사전 학습된 Word2Vec 다운로드

```python
!curl -o GoogleNews-vectors-negative300.bin.gz "https://media.githubusercontent.com/media/mmihaltz/word2vec-GoogleNews-vectors/refs/heads/master/GoogleNews-vectors-negative300.bin.gz?download=true"
```

```python
!pip install gensim
import gensim
```

```python
# 사전 학습된 word2vec을 불러옵니다.
# 로드시 최소 램 8기가 이상이 필요합니다.
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
# 커피와 유사한 단어 찾아보기
model.most_similar(positive=['coffee'], topn=5)
```

```python
# 왕 + 여자 - 남자 = 왕비
model.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
```

```python
# Word2Vec 단어 벡터의 차원 출력하기
len(model['coffee'])
```
